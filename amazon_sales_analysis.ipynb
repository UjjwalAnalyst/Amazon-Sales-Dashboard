```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
#at the end we defined how our date is in the data by not just changing it for example our data:- 25/01/2023
#which meanss day first then month then year/dd-mm-yyyy
```

```python
df['month'] = df['order_date'].dt.month
df['year'] = df['order_date'].dt.year
df['day'] = df['order_date'].dt.day
#These will help you build Power BI charts like Monthly Sales Trend.
```
